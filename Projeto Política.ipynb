```python
## Importando libs

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re
import sweetviz as sv
```

```python
## Requisição html 

html = requests.get('https://indice.legislabrasil.org/public/explorador').content
soup = bs(html, 'html.parser')
```

```python
## Armazenando os nomes dos candidatos

lista_div = soup.find_all('div', class_='box_dep') # Localização dos nomes no html
x = 0
nomes = []

for x in range(len(lista_div)):
    nomes_candidatos = lista_div[x].contents[5].text # Encontrando o texto com nome do candidato
    nomes.append(str(nomes_candidatos)) # Colocando os nomes na lista
nomes    
```

```python
## Armazenando as estrelas dos candidatos

lista_div = soup.find_all('div', class_='ranking') # Localização das estrelas no html
i = 0
y = 0
estrelas = []

for i in range(len(lista_div)): # Loop que passa de candidatao em candidato
    a = 0
    for y in range(len(lista_div[0])): # Loop para verificar as estrelas
        if 'font_ativo' in str(lista_div[i].contents[y]): # Se a estrela está preenchida, adicione 1
            a = a+1
        else:
            None
    estrelas.append(a) # Colocando a soma das estrelas de cada candidato numa lista

estrelas
```

```python
## Armazenando uf e partido dos candidatos que estão no mesmo local do html

lista_div = soup.find_all('div', class_ = 'box_dep') # Localização no html
x = 0
info = []
uf = []
partido = []

for x in range(len(lista_div)):
    info_candidatos = lista_div[x].contents[7].text # Pegando o texto uf e partido
    info.append(str(info_candidatos)) # Armazenando
    uf.append(info[x][29:31]) # Selecionando somente uf e armazenando em lista específica
    partido.append(info[x][60:]) # Selecionando somente partido e armazenando em lista específica
uf
```

```python
## Tratando o partido que veio "sujo"

i = 0
partido_tratado = []

for i in range(len(partido)):
    partido_tratado.append(partido[i].replace('\n ', '').replace(' ', '', 25)) # Retirando info de cada str da lista
partido_tratado
```

```python
## Armazenando o link específico de cada candidato
# Talvez eu use mais tarde

lista_links = soup.find_all('div', class_ = 'box_dep')
x = 0
links = []

for x in range(len(lista_links)):
    link_get = lista_links[x].contents[3].get('href') # Buscando os links dentro da lista
    links.append(str(link_get)) # Armazenando os links
links
```

```python
def buscar_info(url):
    
    html1 = requests.get(url).content
    soup1 = bs(html1, 'html.parser')
    
    ## Anos de mandato
    
    lista_div = soup1.find_all('div', class_ = 'ico_tempo')

    anos.append(str(lista_div[0].text).replace('\n', '').replace('Ano', '').replace(' ', '', 28).replace('s',''))

    ## Em exercício / titular


    lista_h4 = soup1.find_all('h4')

    aux = lista_h4[0].contents[4]
    exer.append(aux[39:42])

    aux1 = lista_h4[0].contents[2]
    titular.append(aux1[25:33].replace('\n', ''))

    ## Gastos de gabinete


    lista_div = soup1.find_all('div', class_ = 'box_3 col_4')
    gasto.append(lista_div[0].contents[3].text.replace('R$ ', '').replace('.', '').replace(',', '.'))

    ## Demais indicadores


    lista_strong = soup1.find_all('strong', class_ = 'cor1')
    
    prod_legis.append(lista_strong[0].text)
    
    fisc.append(lista_strong[1].text)
    
    mob.append(lista_strong[2].text)
    
    alin_part.append(lista_strong[3].text)
    
    return None
```

```python
## Zerando listas usadas para teste

anos = []
exer = []
titular = []
gasto = []
prod_legis = []
fisc = []
mob = []
alin_part = []
```

```python
## Entrando em cada link de candidato para buscar info
## DEMORA EM TORNO DE 10MIN PARA RODAR
t = 0
for t in range(len(links)):
    buscar_info(links[t])
    print(f'{t} Ok')
print('Tudo pronto')
```

```python
## Criando dataframe vazio

df_html = pd.DataFrame()
```

```python
## Adicionando as colunas no df

df_html['nome_candidato'] = nomes
df_html['uf'] = uf
df_html['partido'] = partido_tratado
df_html['estrelas_legisla_int'] = estrelas
df_html['tempo_mandato'] = anos
df_html['em_exercicio'] = exer
df_html['tipo'] = titular
df_html['gasto_gabinete'] = gasto
df_html['producao_legislativa'] = prod_legis
df_html['fiscalizacao'] = fisc
df_html['mobilizaca'] = mob
df_html['alinhamento_partidario'] = alin_part
```

```python
## Percebi que o servidor fica caindo e atrapalha o código, então vou deixar o df independente

df_legisla = df_html.copy() # copiando o df criado a cima
df_legisla.head() # Agora podemos usar o df sem interferência do servidor
```
