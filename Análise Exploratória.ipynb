```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("legisla_brasil.csv", index_col=0)
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.set_theme(style = 'whitegrid')

_ = sns.boxplot(df['gasto_gabinete'])
```

```python
df.groupby('partido')['estrelas_legisla_int',
                      'tempo_mandato',
                      'gasto_gabinete',
                      'producao_legislativa',
                      'fiscalizacao',
                      'mobilizaca',
                      'alinhamento_partidario'].mean().sort_values(by='estrelas_legisla_int', ascending = False)
```

```python
df.groupby('uf')['estrelas_legisla_int',
                      'tempo_mandato',
                      'gasto_gabinete',
                      'producao_legislativa',
                      'fiscalizacao',
                      'mobilizaca',
                      'alinhamento_partidario'].mean().sort_values(by='estrelas_legisla_int', ascending = False)
```

```python
_ = sns.relplot(x = 'alinhamento_partidario', y = 'estrelas_legisla_int', data = df)

## Podemos ver que separadamente, os pilares não tem correlação com as estrelas
```

```python
aux = []
i = 0 

for i in range(len(df)):
    aux.append((df.iloc[i,-4]+df.iloc[i,-3]+df.iloc[i,-2]+df.iloc[i,-1])/4)
```

```python
df['med_pilares'] = aux

df.head()
```

```python
_ = sns.relplot(x = 'med_pilares', y = 'estrelas_legisla_int', data = df)
```

```python
df.groupby('partido')['estrelas_legisla_int',
                          'tempo_mandato',
                          'gasto_gabinete',
                          'producao_legislativa',
                          'fiscalizacao',
                          'mobilizaca',
                          'alinhamento_partidario'].sum().sort_values(by='gasto_gabinete', ascending = False)
```

```python
_ = plt.figure(figsize=(10,7))
_ = df.groupby('uf')['gasto_gabinete'].sum().plot(kind='bar')
_ = plt.title('Gasto total por estado')
```

```python
_ = plt.figure(figsize=(10,7))
_ = df.groupby('partido')['gasto_gabinete'].sum().plot(kind='bar')
_ = plt.title('Gasto total por partido')
```

```python
_ = plt.figure(figsize=(10,7))
_ = df.uf.value_counts().plot(kind='bar')
_ = plt.title('Políticos por estado')
```

```python
_ = plt.figure(figsize=(10,7))
_ = df.partido.value_counts().plot(kind='bar')
_ = plt.title('Políticos por partido')
```
